```python
# Importing the Dataset
from google.colab import drive
import zipfile
import os
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
yList = [] #loss
xList = [] #batch size

with open('/content/drive/My Drive/loss.txt', 'r') as f: 
  for line in f:
    line = f.readline()
    spl = line.split()
    epoch = int(spl[0][1:2])
    if epoch == 2:
      xVal = int(spl[1][:-1]) + 12500
    else:
      xVal = int(spl[1][:-1])
    yList.append(float(spl[3]))
    xList.append(xVal)
    #print(spl)
plt.xlabel("Total Images Trained")
plt.ylabel("Loss Value")

plt.scatter(xList, yList,s=1, alpha=0.5)
plt.show()
#add 12500 to epoch 2
#label
```
